```python
# Mean Squared Error Function

def mse(observations, estimates):
    
    '''
    Input:
        observations - numpy array of values indicating observed values
        estimates - numpy array of values indicating an estimate of values
    Output:
        Mean Square Error
    '''
    # check argument types
    #assert type(observations) == type(np.array([])), "'observations' must be a numpy array"
    #assert type(estimates) == type(np.array([])), "'estimates' must be a numpy array"
    # check length of arrays equal
    assert len(observations) == len(estimates), "Arrays must be of equal length"
    
    # calculations
    difference = observations - estimates
    sq_diff = difference ** 2
    mse = sum(sq_diff)
    
    return mse
```

```python
# Test Train Split

test_size = 100

smooth_1 = estimated_residual_1M[50:-50]
smooth_2 = estimated_residual_2M[50:-50]
smooth_3 = estimated_residual_3M[50:-50]

train_1 = smooth_1[:-test_size]
train_2 = smooth_2[:-test_size]
train_3 = smooth_3[:-test_size]

# test sets
test_1 = smooth_1[-test_size:]
test_2 = smooth_2[-test_size:]
test_3 = smooth_3[-test_size:]
```

```python
# Creating smoothed models

from statsmodels.tsa.api import ExponentialSmoothing
```

```python
#Series 1

model_1A = ExponentialSmoothing(train_1,damped=True,
                              trend="additive",
                              seasonal="additive",
                              seasonal_periods=5).fit(optimized=True)

model_1B = ExponentialSmoothing(train_1,
                              trend="multiplicative",
                              seasonal="multiplicative",
                              seasonal_periods=5).fit(optimized=True)

model_1C = ExponentialSmoothing(train_1,
                              trend="multiplicative",
                              seasonal=None,
                              seasonal_periods=None).fit(optimized=True)

preds_1A = model_1A.forecast(len(test_1))
preds_1B = model_1B.forecast(len(test_1))
preds_1C = model_1C.forecast(len(test_1))
```

```python
# Calculate Error Metric

mse_1A = mse(test_1, preds_1A)
mse_1B = mse(test_1, preds_1B)
mse_1C = mse(test_1, preds_1C)

print("MSE for smooth_1: {:.7}".format(mse_1A))
print("MSE for smooth_2: {:.7}".format(mse_1B))
print("MSE for smooth_3: {:.7}".format(mse_1C))
```
